```python
import pandas as pd
```

```python
df = pd.read_excel("Крылов_тестовое_парсинг.xlsx",sheet_name="Задание 2")
df
```

```python
#Период анализа - с 1.04.23 по 10.04.23, только продажи изделий с общей суммой и себестоимостью больше 0 рублей
df_1 = df[(df['ДатаПродажи'] > '2023-04-01 00:00:00') & (df['ДатаПродажи'] < '2023-04-10 23:59:59') & (df['Себестоимость'] > 0) & (df['ОбщаяСумма'] > 0)]
df_1['Наценка'] = df_1['ОбщаяСумма'] - df_1['Себестоимость']
df_1
```

```python
gold_mean = df_1[df_1['ТоварнаяГруппа'].str.contains("СИ") == False]['Наценка'].mean().round(2)
silver_mean = df_1[df_1['ТоварнаяГруппа'].str.contains("СИ") == True]['Наценка'].mean().round(2)

print('Средняя наценка на ювелирные изделия из золота:',gold_mean,\
    '\nСредняя наценка на ювелирные изделия из серебра:',silver_mean)
```

```python
with open('Средняя наценка на ювелирные изделия.txt', 'w') as file:   
    file.write(f'Средняя наценка на ювелирные изделия из золота: {gold_mean}\n')
    file.write(f'Средняя наценка на ювелирные изделия из серебра: {silver_mean}')
```
